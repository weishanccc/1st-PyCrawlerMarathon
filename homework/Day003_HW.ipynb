```python
# 2. 請取出每一個地區所記錄的第一個時間點跟溫度

'''
Your Code
'''
TimenTemp= dict()
for location in d['cwbopendata']['dataset']['locations']['location']:
    locationName = location['locationName']
    for weatherElement in location['weatherElement']:
        if weatherElement['description'] == '溫度':
            time = weatherElement['time'][0]['dataTime']
            value = weatherElement['time'][0]['elementValue']['value']
            TimenTemp[locationName] = (time,value)

for locationName in TimenTemp:
    print('地區：', locationName, '\n' + '時間與溫度：', TimenTemp[locationName], '\n')
```
